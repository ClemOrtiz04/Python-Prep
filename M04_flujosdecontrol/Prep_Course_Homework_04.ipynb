6) Utilizar un ciclo while para realizar el factorial de un número guardado en una variable, sólo si la variable contiene un número entero mayor a 0

```python
num = 4
factorial = 1
if num == 0:
    print("El factorial de 0 es: 1")
elif num < 0:
    print("Los numeros negativos no tienen factorial")
else:
    while num > 1:
        factorial = factorial * num
        num = num - 1
    print(factorial)
```

7) Crear un ciclo for dentro de un ciclo while

```python
while num < 10:
    print("Tabla del ", num)
    for i in range(1,11):
        print(num, " * ", i, " = ", num * i)
    num = num + 1
```

9) Imprimir los números primos existentes entre 0 y 30

```python
numero = 2
primo = True
while numero <= 30:
    for div in range(2, numero):
        if numero % div == 0:
            primo = False
    if(primo):
        print(f'El número {numero} es un numero primo')
    else:
        primo = True
    numero +=1
```

10) ¿Se puede mejorar el proceso del punto 9? Utilizar las sentencias break y/ó continue para tal fin

```python
numero = 2
rangomax = 30
primo = True
while numero <= rangomax:
    for div in range(2, numero):
        if numero % div == 0:
            primo = False
            break
    if(primo):
        print(f'El número {numero} es un numero primo')
    else:
        primo = True
    numero +=1
```

11) En los puntos 9 y 10, se diseño un código que encuentra números primos y además se lo optimizó. ¿Es posible saber en qué medida se optimizó?

```python
numero = 2
rangomax = 30
primo = True
ciclosbreak = 0
cicloswbreak = 0
#Ciclo donde se analiza la cantidad de ciclos con break
while numero <= rangomax:
    for div in range(2, numero):
        ciclosbreak += 1
        if numero % div == 0:
            primo = False
            break
    if(primo):
        print(f'El número {numero} es un numero primo')
    else:
        primo = True
    numero +=1

#Cantidad de ciclos sin break
numero = 2
while numero <= rangomax:
    primo = True
    for div in range(2, numero):
        cicloswbreak += 1
        if numero % div == 0:
            primo = False
    numero +=1

#Realizar la diferencia de ciclos
print(f'La cantidad de iteraciones ahorradas es en el rango [2,30] es:  {cicloswbreak - ciclosbreak}')
```
